# HDF5

```python
import wradlib as wrl
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
try:
    get_ipython().run_line_magic("matplotlib inline")
except:
    plt.ion()
```

```python
import os

# Set the WRADLIB_DATA environment variable
os.environ['WRADLIB_DATA'] = 'wradlib_data'
```

```python
fpath = "scan-sidpol-120km-14_20001_20240814000001_00.h5"
f = wrl.util.get_wradlib_data_file(fpath)
fcontent = wrl.io.read_generic_hdf5(f)
```

```python
print(fcontent.keys())
```

```python
print(fcontent["where"])
print(fcontent["how"])
print(fcontent["scan0/moment_0"])
print(fcontent["scan0/moment_0"]["attrs"])
print(fcontent["scan0/moment_0"]["data"].shape)
```

```python
fig = plt.figure(figsize=(10, 10))
da = wrl.georef.create_xarray_dataarray(
    fcontent["scan0/moment_0"]["data"]
).wrl.georef.georeference()
im = da.wrl.vis.plot(fig=fig, crs="cg")
```

# GAMIC HDF5

```python
fpath = "scan-sidpol-120km-14_20001_20240814000001_00.h5"
f = wrl.util.get_wradlib_data_file(fpath)
data, metadata = wrl.io.read_gamic_hdf5(f)
```

```python
print(metadata.keys())
print(metadata["VOL"])
print(metadata["SCAN0"].keys())
```

```python
print(data["SCAN0"].keys())
print(data["SCAN0"]["PHIDP"].keys())
print(data["SCAN0"]["PHIDP"]["data"].shape)
```
